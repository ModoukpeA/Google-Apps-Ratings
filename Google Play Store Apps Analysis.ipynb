## Data Cleaning | Imputation | Manipulation

In this part of the notebook, the dataset will be cleaned and null values as well as non numerical values will be corrected.

###  Removing columns that are 90% empty

If there is a column that is empty at 90%, there is no need to fill it. Rather, it should be deleted for better insights.

```python
limit = len(data)* 0.1 #(10% of my rows = 10840)
limit
```

```python
data.dropna(thresh=limit, axis=1, inplace=True)
```

```python
data.shape
```

The number of column has not changed; that means that all the columns have null values that are less than 90% of the dataset content.

### Filling null values with the appropriate value (mode, median or mean)

#### Define a function impute_median

```python
def impute_median(series):
    return series.fillna(series.median())
```

#### Apply it to the **Rating** column

```python
data.Rating = data['Rating'].transform(impute_median) 
```

Let's check whether the null values of this column have been filled.

```python
data.isnull().sum()
```

There are no more null values in the **Rating** column.

The other columns that contain null values are : **Type**, **Current Vers** and **Android Vers**.
    
But as they are categorical values, they will be filled by their respective mode values.

```python
data['Type'].fillna(str(data['Type'].mode().values[0]), inplace=True)
data['Current Ver'].fillna(str(data['Current Ver'].mode().values[0]), inplace=True)
data['Android Ver'].fillna(str(data['Android Ver'].mode().values[0]), inplace=True)
```

```python
data.isnull().sum()
```

All the null values are now filled.

Now, there are some **cleaning and manipulation** tasks that need to be performed on some columns. Let's have a quick look at a set of the data.

```python
data.tail()
```

### Converting some columns into numerical columns

According to the previous analysis, only the **Rating** column is numerical. But why ? 

Why not **Reviews** ?

Why not **Price** ?

or **Installs** ?

```python
data['Price'] = data['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
data['Price'] = data['Price'].apply(lambda x: float(x))
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')
data['Installs'] = data['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x))
data['Installs'] = data['Installs'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else str(x))
data['Installs'] = data['Installs'].apply(lambda x: float(x))
```

```python
data.tail()
```

Let's check if we now have more than one numerical column.

```python
data.hist()
```

The four (04) numerical columns are present.

## Data Visualization

There is more than 10000 records in the dataset. To make visualizations, we need to group the records by category.

```python
grouped_data = data.groupby('Category')
ratings = grouped_data['Rating'].agg(np.mean)
price = grouped_data['Price'].agg(np.sum)
installs = grouped_data['Installs'].agg(np.mean)
reviews = grouped_data['Reviews'].agg(np.mean)
```

```python
plt.figure(figsize=(20,8))
plt.plot(ratings,'g^', color='g')
plt.xticks(rotation=100)
plt.title('Applications Ratings')
plt.xlabel('Categories')
plt.ylabel('Ratings')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(price,'ro', color='pink')
plt.xticks(rotation=100)
plt.title('Applications Prices')
plt.xlabel('Categories')
plt.ylabel('Prices')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(installs,'r--', color='red')
plt.xticks(rotation=100)
plt.title('Applications Installs')
plt.xlabel('Categories')
plt.ylabel('Installs')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(reviews,'--', color='black')
plt.xticks(rotation=100)
plt.title('Applications Reviews')
plt.xlabel('Categories')
plt.ylabel('Reviews')
plt.show()
```
